# The First Analysis

## Objective:

```python
# the variables for the second analysis I want to choose 
columns_to_keep = ['DEMO_age','CONNECTION_social_num_close_friends','CONNECTION_preference_time_family','CONNECTION_social_days_family_p7d','CONNECTION_preference_time_friends','CONNECTION_social_days_friends_p7d','CONNECTION_social_days_neighbours_p7d', 'CONNECTION_preference_time_neighbours','WELLNESS_self_rated_mental_health']  

# keep the columns I want  
data = df[columns_to_keep]

# delet the missing values
import numpy as np
data = data.dropna()
data
```

# The Second Analysis

## Objective:
I want to know whether these variables have relationship with the WELLNESS_self_rated_mental_health (the reason why I choose this one rather than the identity of mental health or the satidfaction is that these two have too much missing values which will cause the result be not that presice) and which one has the greater impact on the WELLNESS_self_rated_mental_health. 

```python
data = data.drop(data[data['WELLNESS_self_rated_mental_health'] == "Presented but no response"].index)
data
```

```python
# encoding the dependent variable

from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LogisticRegression  
from sklearn.preprocessing import LabelEncoder  
from sklearn.metrics import classification_report, confusion_matrix

# create LabelEncoder  
label_encoder = LabelEncoder()  
  
# Coding the dependent variable  
data['WELLNESS_self_rated_mental_health_encoded'] = label_encoder.fit_transform(data['WELLNESS_self_rated_mental_health'])  
  
# View the encoded category  
print(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
```

```python
data
```

```python
del data['WELLNESS_self_rated_mental_health']
data
```

```python
# choose the data
X = data.iloc[:, :8]
Y = data.iloc[:, 8]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
X_train
```

```python
X.iloc[:,0] = (data.iloc[:,0]-min(data.iloc[:,0]))/max(data.iloc[:,0])
X.iloc[:,1] = (data.iloc[:,1]-min(data.iloc[:,1]))/max(data.iloc[:,1])
X.iloc[:,2] = (data.iloc[:,2]-min(data.iloc[:,2]))/max(data.iloc[:,2])
X.iloc[:,3] = (data.iloc[:,3]-min(data.iloc[:,3]))/max(data.iloc[:,3])
X.iloc[:,4] = (data.iloc[:,4]-min(data.iloc[:,4]))/max(data.iloc[:,4])
X.iloc[:,5] = (data.iloc[:,5]-min(data.iloc[:,5]))/max(data.iloc[:,5])
X.iloc[:,6] = (data.iloc[:,6]-min(data.iloc[:,6]))/max(data.iloc[:,6])
X.iloc[:,7] = (data.iloc[:,7]-min(data.iloc[:,7]))/max(data.iloc[:,7])
```

```python
X
```

```python
# Create a logistic regression model object and train the model 
model = LogisticRegression(max_iter=1000)  
model.fit(X_train, y_train)  
  
# Make predictions using the test set  
y_pred = model.predict(X_test)  
  

print(confusion_matrix(y_test, y_pred))  
```

```python
model.coef_
```

Result: for Group 3, we can know that the social days for neighbor impact more about the mental health.

# The Third Analysis

## Objective: to get the better model for analysising

The variables are same as the the second analysis. 

```python
from sklearn import tree 

# DecisionTreeClassifier
clf = tree.DecisionTreeClassifier()


clf = clf.fit(X_train, y_train)

# Make predictions using the test set 
y_pred = clf.predict(X_test)  
    
print(confusion_matrix(y_test, y_pred))  
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 1000)

clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)  
   
print(confusion_matrix(y_test, y_pred))  
```
